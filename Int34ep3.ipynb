```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/dsrscientist/IBM_HR_Attrition_Rate_Analytics/master/ibm-hr-analytics-employee-attrition-performance.zip")
df
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='magma')
plt.title('Correlation Heatmap', fontsize=30)
plt.show()
```

```python
sns.countplot('Attrition', data=df)
plt.title('Attrition count', fontsize=26)
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
categorical_cols = [feature for feature in df.columns if df[feature].dtypes=='object']
df[categorical_cols].sample(10)
```

```python
numerical_cols = [feature for feature in df.columns if df[feature].dtypes!='object']
df[numerical_cols].sample(10)
```
